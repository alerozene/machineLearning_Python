## Feedforward fitting equation and number of layers
The most basic equation for a neural network is the one-layer feedforward equation, which reads:
\begin{equation}
\hat{y} = \sigma(W(x))
\end{equation}
where $\hat{y}$ is the estimated output, $W$ is the weight and $\sigma$ is the sigmoid equation $1+e^{-x}$

In the case of a two layer neural network, the feedforward fitting is:
\begin{equation}
\hat{y} = \sigma(W_2(\sigma(W_1x)))
\end{equation}
The python implementation:
